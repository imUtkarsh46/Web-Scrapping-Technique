# Main Webscraping

```python
comp = soup.find_all('div', class_='company-content-wrapper')
```

```python
len(comp)
```

```python
name = []
rating = []
reviews = []
ctype = []
location = []
old = []
comp_size = []
salary = []
jobs = []    
```

```python
for i in comp:
    name.append(i.find('h2').text.strip())
    rating.append(i.find_all('p', class_='rating')[0].text.strip())
    reviews.append(i.find_all('a', class_='review-count')[0].text.strip())
    ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
    location.append(i.find_all('p', class_='infoEntity')[1].text.strip())
    old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
    comp_size.append(i.find_all('p', class_='infoEntity')[3].text.strip())
    
```

```python
other = soup.find_all('div', class_='company-action-center')
```

```python
for i in other:
    salary.append(i.find_all('span', class_='caption-subdued-large')[1].text.strip())
    jobs.append(i.find_all('span', class_='caption-subdued-large')[3].text.strip())
    
```

```python
df = {
    'Comp_name' : name,
    'Ratings' : rating,
    'Review_count' : reviews,
    'Type': ctype,
    'HQ_Loc' : location,
    'Years_old' : old,
    'Comp_Head_count': comp_size,
    'Salary' : salary,
    'Jobs' : jobs
}
```

```python
data = pd.DataFrame(df)
```

# Extract Entire Data Now

```python
import numpy as np

final=pd.DataFrame()
for j in range(1,501):
  webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('div',class_='company-content-wrapper')
  other = soup.find_all('div', class_='company-action-center')

  name=[]
  rating=[]
  reviews=[]
  ctype=[]
  hq=[]
  how_old=[]
  no_of_employee=[]
  salary = []
  jobs = []

  for k in other:
    try:
        salary.append(k.find_all('span', class_='caption-subdued-large')[1].text.strip())
    except:
        salary.append(np.nan)
            
    try:    
        jobs.append(k.find_all('span', class_='caption-subdued-large')[3].text.strip())
    except:
        jobs.append(np.nan)
    

  for i in company:

    try:
       name.append(i.find('h2').text.strip())
    except:
       name.append(np.nan)

    try:
       rating.append(i.find('p',class_='rating').text.strip())
    except:
       rating.append(np.nan)
   
    try:

      reviews.append(i.find('a' , class_='review-count').text.strip())
    except:
      reviews.append(np.nan)

    try:

      ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    except:
      ctype.append(np.nan)
    try:

      hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)
    
    try:

      how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    except:
      how_old.append(np.nan)
    try:
      no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
      no_of_employee.append(np.nan)
    

  df=pd.DataFrame({'name':name,
    'rating':rating,
    'reviews':reviews,
    'company_type':ctype,
    'Head_Quarters':hq,
    'Company_Age':how_old,
    'No_of_Employee':no_of_employee,
    'Salary':salary,
    'jobs':jobs
    })
  
  final=final.append(df,ignore_index=True)
```
